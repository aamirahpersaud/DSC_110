```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

This part of the code imports the pandas library, given the alias pd to import the matplotlib.pyplot module to create animated visualizations under "pit." The performance of the classification model is then evaluated.

```python
X_train = pd.read_csv('https://raw.githubusercontent.com/osoliman/DSC110/refs/heads/main/Datasets/X_train.csv')
X_test = pd.read_csv('https://raw.githubusercontent.com/osoliman/DSC110/refs/heads/main/Datasets/X_test.csv')
y_train = pd.read_csv('https://raw.githubusercontent.com/osoliman/DSC110/refs/heads/main/Datasets/y_train.csv')
y_test = pd.read_csv('https://raw.githubusercontent.com/osoliman/DSC110/refs/heads/main/Datasets/y_test.csv')

#to prevent warning, converting the y from df to 1d series
y_train = y_train.squeeze()
y_test  = y_test.squeeze()
```

This code loads the training and testing datasets for x and y as the target variables from the CSV files. These tests and trains are converted into pandas series from DataFrames with squeeze.

```python
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
```

This creates an object for the model, making it easier to do so.

```python
from sklearn.metrics import classification_report, roc_auc_score

y_pred = rf.predict(X_test)
y_prob = rf.predict_proba(X_test)[:, 1]

print(classification_report(y_test, y_pred))
#print(classification_report(y_test, y_pred, labels=[1]))
print(f"AUC-ROC: {roc_auc_score(y_test, y_prob):.4f}")
```

This scores and evaluates the performance of the model. These two imported functions provide accuracy, sensitivity, and precision. This makes predictions of the model based on the threshold. Yes or No are being predicted to provide an exact number between 0 and 1. The metrics are provided based on the classes. The specificity can be determined by the sensitivity value. The AUC-ROC number can be used to represent the percentage range.

This array contains the predicted probabilities that belongs to a certain class in each instance. This assures the model's confidence. If closer to 1, the model is more confident than it would be if the predicted value is less close to 1.
