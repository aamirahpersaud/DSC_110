```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

This code loads the training and testing datasets for x and y as the target variables from the CSV files. These tests and trains are converted into pandas series from DataFrames with squeeze. The first two lines of the code establish the independent variables for training and testing into pandas. The next two lines of the code then load the dependent variables. The corresponding labels are providing. The DataFrames are then converted into the pandas series.

```python
X_train = pd.read_csv('https://raw.githubusercontent.com/osoliman/DSC110/refs/heads/main/Datasets/X_train.csv')
X_test = pd.read_csv('https://raw.githubusercontent.com/osoliman/DSC110/refs/heads/main/Datasets/X_test.csv')
y_train = pd.read_csv('https://raw.githubusercontent.com/osoliman/DSC110/refs/heads/main/Datasets/y_train.csv')
y_test = pd.read_csv('https://raw.githubusercontent.com/osoliman/DSC110/refs/heads/main/Datasets/y_test.csv')

#to prevent warning, converting the y from df to 1d series
y_train = y_train.squeeze()
y_test  = y_test.squeeze()
```

This creates an object for the model. The "RandomForestClassifier" builds several decision trees during the training part of the set. It learns the relationship between the x and y train. The "random_state" part of the code controls the randomness of the model to ensure that the same exact results are ran regardless of how many times the code is ran.

```python
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
```

This query evaluates the trained random Forest Classifier where it uses the model to predict the class probabilities and predictions. The predictions are then used with the true labeks in order to calculate the AUC-ROC score. This curve plots the sensitivity against the specificity. Specificity measures the true negative rate while sensitivity measures the true positive rate. "AUC-ROC" represents the ability of the model to distinguish between classses where the score indicates where the model correctly ranks in terms of a random positive instance above a random negative one.
(Reflected on based on commentary)


```python
from sklearn.metrics import classification_report, roc_auc_score

y_pred = rf.predict(X_test)
y_prob = rf.predict_proba(X_test)[:, 1]

print(classification_report(y_test, y_pred))
#print(classification_report(y_test, y_pred, labels=[1]))
print(f"AUC-ROC: {roc_auc_score(y_test, y_prob):.4f}")
```
