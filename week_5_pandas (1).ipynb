Data Cleaning

```python
#Here is our comment
df.isna().sum()
```

This calculates the total number of missing values, for each column in the DataFrame, helping to identify which columns have data gaps.

```python
df.isna().mean()*100
```

 This calculates the percentage of missing values for each column, emphasizing the missing data.

```python
df_clean = df.dropna(subset=['has_diabetes']).copy()
```

This creates a new DataFrame called df_clean by removing any rows from the original df where the 'has_diabetes' column has a missing value. This ensures df_clean is a completely independent copy.

```python
df_clean.isna().mean() *100
```

This confirms the percentage of missing values in the new df_clean DataFrame to ensure that 'has_diabetes' now has 0% missing values and to see how other columns were affected.

```python
df_clean.shape
```

This confirms how many rows were previously removed.

```python
mean_bp = df_clean['systolic_bp']. mean()
df_clean['systolic_bp'] = df_clean['systolic_bp'].fillna(mean_bp)
```

```python
mean_bp = df_clean['systolic_bp'].mean()
print (mean_bp)
```

Transform Variables

```python
#Converting numerical to categorical
# Create BMI categories
df['bmi_category'] = pd.cut(df['bmi'],
                            bins=[0, 18.5, 25, 30, 100],
                            labels=['Underweight', 'Normal', 'Overweight', 'Obese'])
```

This shows how to create new categorical features from existing numerical categories. It defines the BMI categories based on numerical BMI ranges using the pd.cut()

```python
df['bmi_category'].head()
```

The head and value counts are displayed from the newly created bmi_category column





```python
df['bmi_category'].value_counts()
```

Categorical Variable Transformation


```python
df['education'].head()
```

```python
education_map = {
    1: 'Less than 9th Grade',
    2: '9-11th Grade',
    3: 'High School Grad',
    4: 'Some College or AA Degree',
    5: 'College Graduate',
    9: 'Do Not Know',
    'Nan': 'Missing'

}
```

This shows how to change numerical categorical codes into descriptive labels. It defines the education_map to map numerical education codes to descriptives.

```python
education_map[1]
```

```python
df['education_label'] = df['education'].map(education_map)
```

```python
df['education_label'].value_counts()
```

```python
df['education_label']
```

The education labels are the descriptions that the numerical does apply to. The dictionary education_map displays the numerical education codes to descriptions. This applies the map to this education column and shows the value counts.
